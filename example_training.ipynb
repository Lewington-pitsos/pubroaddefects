# Example Training Process

Here we give an example of how to use this repository to create a dataset and then begin training a tensorflow model with it. Before you begin here it is assumed that you have extracted raw data as described in `README.md`. Having done this, the steps are:

1. Create the dataset form the raw annotated images
2. Createa TFDataset to handle the loading of the dataset
3. Train a tensorflow model on that data

Once this finishes there should be a whole bunch of `.record` files sitting in `data/processed/256-public_defects_big5_subset/`. There should also be one `meta.json` file that contains important information about the datset. Now all we need to do is load them into a TFDataset.

```python
import json

import tensorflow as tf
import matplotlib.pyplot as plt

from app.glob import *

from app.assist import full_contents
```

Define some helper functions for dealing with TFRecords. 

```python
def make_set_shape_fn(input_dims, output_range):
    def fixup_shape(img, label_data):
        img.set_shape(input_dims)
        
        label_data.set_shape(output_range) 

        return img, label_data
    
    return fixup_shape

# The constants below are all from app/glob.py
parse_format = {
    TF_CLASS_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_OBJECT_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_COUNT_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_XMINS_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_XMAXS_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_YMINS_LABEL: tf.io.FixedLenFeature([], tf.string),
    TF_YMAXS_LABEL: tf.io.FixedLenFeature([], tf.string),

    TF_DATASET_NAME: tf.io.FixedLenFeature([], tf.string),
    TF_FILENAME: tf.io.FixedLenFeature([], tf.string),
    
    TF_IMAGE: tf.io.FixedLenFeature([], tf.string),
}

def parse_tfrecord_classonly(record):
    parsed_record = tf.io.parse_single_example(record, parse_format)

    return tf.io.parse_tensor(parsed_record[TF_IMAGE], out_type=tf.float32),\
        tf.io.parse_tensor(parsed_record[TF_CLASS_LABEL], out_type=tf.int16)  
```

Process dataset metadata

```python
data_name = PROCESSED_DATA_PATH + "256-public_defects_big5_subset/"

with open(data_name + "meta.json") as f:
     metadata = json.load(f) 

records_per_file=metadata["records_per_file"] 
final_file_count=metadata["final_file_count"]
final_file_name=metadata["final_file_name"]
        
if final_file_count == 0:
    final_file_count = records_per_file

fix_shape_fn = make_set_shape_fn(metadata["dimensions"], len(metadata["target_list"]))
```

```python
record_filenames = [f for f in full_contents(data_name) if f.split(".")[-1] == "record"]

dataset = tf.data.TFRecordDataset(record_filenames)\
    .map(parse_tfrecord_classonly, num_parallel_calls=tf.data.experimental.AUTOTUNE)\
    .map(fix_shape_fn, num_parallel_calls=tf.data.experimental.AUTOTUNE)\
    .batch(16)
```

```python
img, label = next(iter(dataset)) # acquire single batch
plt.imshow(img[0] / 255) 
```

Now we create a tensorflow model and train it on the dataset we created.

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=metadata["dimensions"]),
  tf.keras.layers.Conv2D(256, (3, 3), activation='relu'),
  tf.keras.layers.Conv2D(128, (2, 2), activation='relu'),
  tf.keras.layers.Conv2D(64, (2, 2), activation='relu'),
#   tf.keras.layers.AveragePooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(len(metadata["target_list"]), activation='sigmoid') # sigmoid, since we're doing multi-label classification
])
model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(0.001),
    metrics=['accuracy'],
)

model.fit(
    dataset,
    epochs=6,
)
```

As you can see, this model performs like the utter garbage it is. In practice you'll probably want to download a pretrained model and start with that.
